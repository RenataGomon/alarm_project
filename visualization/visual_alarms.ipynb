## Distribution of alarms by day of the week

```python
df['start'] = pd.to_datetime(df['start'])
df['day_of_week_num'] = df['start'].dt.weekday

days_uk = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}
df['day_of_week'] = df['day_of_week_num'].map(days_uk)

counts = df['day_of_week'].value_counts()
counts = counts.reindex(sorted(counts.index, key=lambda x: list(days_uk.values()).index(x)))

percentages = counts / counts.sum()

normed_values = (percentages - percentages.min()) / (percentages.max() - percentages.min())

cmap = cm.get_cmap('YlOrRd')  
colors = [cmap(val) for val in normed_values]

plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    colors=colors,
    wedgeprops=dict(width=0.6)
)

def is_dark_color(rgb):
    r, g, b = rgb[:3]
    luminance = 0.299 * r + 0.587 * g + 0.114 * b
    return luminance < 0.5

for autotext, color in zip(autotexts, colors):
    autotext.set_color('white' if is_dark_color(color) else 'black')
    autotext.set_fontsize(12)
    autotext.set_weight("bold")

plt.title('Distribution of alarms by day of the week', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
df['start'] = pd.to_datetime(df['start'])
df['end'] = pd.to_datetime(df['end'])

alarm_counts = df.groupby('region_city').size().reset_index(name='alarm_count')

region_mapping = dict(zip(regions_df['ukr_region_obl'], regions_df['en_region_obl']))
alarm_counts['region_name_en'] = alarm_counts['region_city'].map(region_mapping)

merged = geo_df.merge(alarm_counts, left_on='shapeName', right_on='region_name_en', how='left')

max_val = alarm_counts['alarm_count'].max()
merged['numeric_alarms'] = merged['alarm_count'].fillna(max_val)
merged.loc[merged['region_city'] == 'Крим', 'numeric_alarms'] = max_val

fig, ax = plt.subplots(1, 1, figsize=(12, 10))
norm = Normalize(vmin=0, vmax=max_val)
cmap = cm.get_cmap('YlOrRd')
merged.plot(column='numeric_alarms', cmap=cmap, linewidth=0.8, ax=ax, edgecolor='0.5')

offsets = {
    "Київська обл.": (0, -0.3),
    "Тернопільська обл.": (0, -0.2),
    "Одеська обл.": (0, -0.45),
}

for idx, row in merged.iterrows():
    centroid = row['geometry'].centroid
    dx, dy = offsets.get(row['region_city'], (0, 0)) 

    if pd.isna(row['alarm_count']):
        text = f"{row['region_city']}\nongoing"
    elif row['region_city'] == 'Крим':
        text = f"{row['region_city']}\n{row['alarm_count']:.1f} + \nongoing"
    else:
        text = f"{row['region_city']}\n{row['alarm_count']:.1f}"
    
    color = 'black' if row['numeric_alarms'] < max_val / 2 else 'white'
    
    ax.text(
        centroid.x + dx, centroid.y + dy,
        text,
        ha='center', va='center',
        fontsize=7, fontweight='bold',
        color=color
    )



sm = ScalarMappable(cmap=cmap, norm=norm)
sm._A = []
cbar = fig.colorbar(sm, ax=ax, shrink=0.6, pad=0.02, fraction=0.046)
cbar.set_label('Number of air alerts')

plt.title('Number of air alerts by regions of Ukraine')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Duration of air alerts by regions of Ukraine

```python
df['start'] = pd.to_datetime(df['start'])
df['end'] = pd.to_datetime(df['end'])

df['duration_hours'] = (df['end'] - df['start']).dt.total_seconds() / 3600

alarm_hours = df.groupby('region_city')['duration_hours'].sum().reset_index()
alarm_hours['duration_hours'] = alarm_hours['duration_hours'].round(1) 

region_mapping = dict(zip(regions_df['ukr_region_obl'], regions_df['en_region_obl']))
alarm_hours['region_name_en'] = alarm_hours['region_city'].map(region_mapping)
merged = geo_df.merge(alarm_hours, left_on='shapeName', right_on='region_name_en', how='left')

max_val = alarm_hours['duration_hours'].max()
merged['numeric_hours'] = merged['duration_hours'].fillna(max_val)
merged.loc[merged['region_city'] == 'Крим', 'numeric_hours'] = max_val

fig, ax = plt.subplots(1, 1, figsize=(12, 10))
norm = Normalize(vmin=0, vmax=max_val)
cmap = cm.get_cmap('YlOrRd')
merged.plot(column='numeric_hours', cmap=cmap, linewidth=0.8, ax=ax, edgecolor='0.5')

offsets = {
    "Київська обл.": (0, -0.3),
    "Тернопільська обл.": (0, -0.2),
    "Одеська обл.": (0, -0.45),
}

for idx, row in merged.iterrows():
    centroid = row['geometry'].centroid
    dx, dy = offsets.get(row['region_city'], (0, 0)) 

    if pd.isna(row['duration_hours']):
        text = f"{row['region_city']}\nno data"
    elif row['region_city'] == 'Крим':
        text = f"{row['region_city']}\n{row['duration_hours']:.1f} + \nongoing"
    else:
        text = f"{row['region_city']}\n{row['duration_hours']:.1f}"
    
    color = 'black' if row['numeric_hours'] < max_val / 2 else 'white'
    
    ax.text(
        centroid.x + dx, centroid.y + dy,
        text,
        ha='center', va='center',
        fontsize=7, fontweight='bold',
        color=color
    )

sm = ScalarMappable(cmap=cmap, norm=norm)
sm._A = []
cbar = fig.colorbar(sm, ax=ax, shrink=0.6, pad=0.02, fraction=0.046)
cbar.set_label('Number of hours of airborne alarms')

total_hours = merged['numeric_hours'].sum()
plt.title('Duration of air alerts by regions of Ukraine')
plt.axis('off')
plt.tight_layout()
plt.show()
```
